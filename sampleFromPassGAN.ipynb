# Set me up:

```python
import tensorflow as tf # only 1.14.0 tested
import tensorflow_hub as hub
import numpy as np
from util import *

MODEL_PATH = './DATA/TFHUB_models/BNK_RESNET_PassGAN_2/'
CHARMAP_PATH = './DATA/MISC/char_map.pickle'
LATENT_SIZE = 64
MHUB_SIGNATURE = 'latent_to_data'
```

# Load hub model:

```python
PassGAN = hub.Module(MODEL_PATH)
```

# Generate Passwords:

```python
N_PASSWORDS = 2**8

z = np.random.normal(size=(N_PASSWORDS, LATENT_SIZE))
x = PassGAN(z, signature=MHUB_SIGNATURE)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.tables_initializer())
    _pwd = sess.run(x)
pwd = clean(_pwd)

print(*pwd, sep='\n')
```
